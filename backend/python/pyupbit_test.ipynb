```python
df = pyupbit.get_ohlcv("KRW-BTC", interval="minute1")
```

```python
df
```

```python
# sql
last_time = df.index[-1]
last_time
```

```python
# upbit 시간 업데이트
t = last_time - timedelta(hours=9) + timedelta(minutes=1)
t
```

```python
# 현재 시간
now = datetime.now()  - timedelta(hours=9) + timedelta(minutes=1)
now
```

```python
diff_time = now - t
diff_time
```

```python
minute_counts = int(diff_time.total_seconds()/60)+1
minute_counts
```

```python
pyupbit.get_ohlcv("KRW-BTC", interval="minute1", count=minute_counts, period=0.5)
```
